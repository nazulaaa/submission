## Visualization & Explanatory Analysis

Setelah melakukan analisis data yang melibatkan penggunaan fungsi groupby, langkah selanjutnya adalah menampilkan hasil analisis tersebut menggunakan visualisasi data. Saya menggunakan line chart dan bar chart untuk menggambarkan temuan dari analisis yang telah dilakukan.

### Pertanyaan 1:

```python

workingday_pivot = workingday.reset_index().pivot(index='mnth', columns='workingday', values='cnt')

months_order = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']

plt.figure(figsize=(10, 6))
for category in workingday_pivot.columns:
    plt.plot(workingday_pivot.index, workingday_pivot[category], marker='o', label=category)

plt.title('Penggunaan Sepeda Motor per Bulan Berdasarkan Hari Kerja')
plt.xlabel('Bulan')
plt.ylabel('Jumlah')
plt.xticks(range(1, 13), months_order, rotation=45)
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


max_work = workingday.loc[workingday.index.get_level_values('workingday') == 'work'].max()
min_work = workingday.loc[workingday.index.get_level_values('workingday') == 'work'].min()
max_notwork = workingday.loc[workingday.index.get_level_values('workingday') == 'not work '].max()
min_notwork = workingday.loc[workingday.index.get_level_values('workingday') == 'not work '].min()


categories = ['Max Work', 'Min Work', 'Max Not Work', 'Min Not Work']
values = [max_work['cnt'], min_work['cnt'], max_notwork['cnt'], min_notwork['cnt']]

plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values, color=['blue', 'blue', 'orange', 'orange'])


for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 10, value, ha='center', va='bottom')


plt.title('Jumlah Penggunaan Sepeda Motor Tertinggi dan Terendah Berdasarkan Hari Kerja')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
df['mnth'] = df['mnth'].replace({1: 'Januari',
                                 2: 'Februari',
                                 3: 'Maret',
                                 4: 'April',
                                 5: 'Mei',
                                 6:'Juni',
                                 7:'Juli',
                                 8:'Agustus',
                                 9:'September',
                                 10:'Oktober',
                                 11:'November',
                                 12:'Desember'
                                 })
```

```python
season = df.groupby(by = ['season','mnth']).agg({

  'cnt' : 'sum'

})

season
```

```python


season_totals = df.groupby('season')['cnt'].sum()

max_season = season_totals.idxmax()
min_season = season_totals.idxmin()

plt.figure(figsize=(10, 6))
bars = plt.bar(season_totals.index, season_totals.values, color=['blue', 'green', 'orange', 'red'])

plt.text(max_season, season_totals[max_season], f'{season_totals[max_season]}', ha='center', va='bottom')
plt.text(min_season, season_totals[min_season], f'{season_totals[min_season]}', ha='center', va='bottom')

plt.title('Total Penggunaan Sepeda Motor Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah')


plt.xticks(season_totals.index, ['Musim Semi', 'Musim Panas', 'Musim Gugur', 'Musim Dingin'])

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python


season_month_totals = df.groupby(['season', 'mnth'])['cnt'].sum()


all_months_season_totals = pd.DataFrame(index=range(1, 13))

for season in season_month_totals.index.get_level_values('season').unique():
    df['season'] = df['season'].replace({1: 'Semi',
                                 2: 'Panas',
                                 3: 'Gugur',
                                 4: 'Dingin',})
    all_months_season_totals[f'Musim {season}'] = season_month_totals[season]

plt.figure(figsize=(10, 6))

all_months_season_totals.plot(kind='bar', color=['blue', 'green', 'orange', 'red'])


plt.title('Total Penggunaan Sepeda Motor Berdasarkan Musim')
plt.xlabel('Bulan')
plt.ylabel('Jumlah')


plt.xticks(range(0, 12), ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'], rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.legend(title='Musim')
plt.show()
```
